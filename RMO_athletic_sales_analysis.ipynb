```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
combine2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
combine2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combine_2020_2021_df = pd.concat([combine2020_df, combine2021_df], axis=0) 
combine_2020_2021_df.reset_index(drop=True, inplace=True)
combine_2020_2021_df.head()
```

```python
# Check if any values are null.
null_validation = combine_2020_2021_df.isnull().any()
null_validation
```

```python
# Check the data type of each column
dt_type_21_22 = combine_2020_2021_df.dtypes
data_type_validator = pd.DataFrame({'Non-Null': null_validation, 
                                    'Data Type': dt_type_21_22})
print(data_type_validator)

```

```python
# Convert the "invoice_date" column to a datetime datatype
if 'invoice_date' in combine_2020_2021_df.columns:
    combine_2020_2021_df['invoice_date'] = pd.to_datetime(combine_2020_2021_df['invoice_date'])

combine_2020_2021_df
    

```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
num_prod = combine_2020_2021_df.groupby(["region", "state", "city"])[["units_sold"]].sum()
# Rename the sum to "Total_Products_Sold".
num_prod = num_prod.rename(columns={"units_sold":"Total_Products_Sold"})
#sort by decending 
num_prod = num_prod.sort_values(by="Total_Products_Sold", ascending=False)
# Show the top 5 results.
num_prod.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
no_col = pd.pivot_table(combine_2020_2021_df, values='units_sold', 
                                            index= ['region', 
                                                    'state', 
                                                    'city'],
                                                    aggfunc='sum')
# Rename the "units_sold" column to "Total_Products_Sold"
no_col = no_col.rename(columns={"units_sold":"Total_Products_Sold"})
#sort by decending as shown in the picture
no_col = no_col.sort_values(by="Total_Products_Sold", ascending=False)
# Show the top 5 results.
num_prod.head(5)
```

```python
# Show the total sales for the products sold for each region, state, and city.
tot_sale = combine_2020_2021_df.groupby(["region", 
                                         "state", 
                                         "city"])[["total_sales"]].sum()
# Rename the "total_sales" column to "Total Sales"
tot_sale = tot_sale.rename(columns={"total_sales":"Total Sales"})
#sort by decending 
tot_sale = tot_sale.sort_values(by="Total Sales", ascending=False)
# Show the top 5 results.
tot_sale.head(5)
```

### 4. Determine which Retailer had the Most Sales

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
sale_data = combine_2020_2021_df[combine_2020_2021_df["product"].str.contains("Women's Athletic Footwear")]  
sale_data
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.  
no_sale_w = sale_data.groupby(["retailer",
                                "region", 
                                "state", 
                                "city"])[["units_sold"]].sum()
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
no_sale_w = no_sale_w.rename(columns={
                                    "units_sold":"Womens_Footwear_Units_Sold"
                                    })
no_sale_w = no_sale_w.sort_values(by="Womens_Footwear_Units_Sold", 
                                  ascending=False)

# Show the top 5 results.
no_sale_w.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
no_sale_w_alt = pd.pivot_table(sale_data, values='units_sold', 
                                            index= ['retailer',
                                                    'region', 
                                                    'state', 
                                                    'city'],
                                                    aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
no_sale_w_alt = no_sale_w_alt.rename(columns={
                                        "units_sold":"Womens_Footwear_Units_Sold"
                                        })
#sort by decending 
no_sale_w_alt = no_sale_w_alt.sort_values(by="Womens_Footwear_Units_Sold", 
                                                ascending=False)
# Show the top 5 results.
no_sale_w_alt.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
inv_sale = pd.pivot_table(sale_data, values='total_sales', 
                                        index= ['invoice_date'],
                                                aggfunc='sum')

# Optional: Rename the "total_sales" column to "Total Sales"
inv_sale = inv_sale.rename(columns={"total_sales":"Total Sales"})

#sort
#inv_sale = inv_sale.sort_values(by="Total Sales", ascending=False)

# Show the table.
inv_sale.head(10)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
#Daily bin with resample method
daily_sample = inv_sale.resample('D').sum()
# Sort the resampled pivot table in descending order on "Total Sales".
daily_sample = daily_sample.sort_values(by="Total Sales", ascending=False)

daily_sample.head(10)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
#Weekly bin with resample method
wkly_sample = inv_sale.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
wkly_sample = wkly_sample.sort_values(by="Total Sales", ascending=False)

wkly_sample.head(10)
```
